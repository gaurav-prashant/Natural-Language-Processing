### Stemming in NLP
Stemming chops off parts of a word like (ing,ed,s) to get it's stem(root).
It does not guarantee a real dictionary word - just common base form.

Eg:
- Running-----Run
- Runs------- Run
- Easily --------- Easili
- Studies----------Studi

* Notice how "easily--> easili and "Studies" are not valid english words -- that's normal in stemming

```python
from nltk.stem import PorterStemmer
stemmer = PorterStemmer()
word = ["Running","Runs","Change","Studies"]
for i in word:
    print(i, "<----------->",stemmer.stem(i))
```

##### Lemmatization
More accurate then stemming

```python
from nltk.stem import WordNetLemmatizer
from nltk.corpus import wordnet

nltk.download("wordnet") #--> WordNet is a large dictionary of english word with meaning and alternate words
nltk.download("omw-1.4") #-->updated english data

#Initiaize the Lemmatizer
lem = WordNetLemmatizer()
w = ["Running","Runs","Change","Studies"]
for i in w:
     print(i, "<----------->",lem.lemmatize(i))
```

```python
t = ["I love NLP üòç! It‚Äôs amazing to analyze 1000s of messages, emails, and tweets!!!",
    "Cleaning text data isn't easy üòÖ ‚Äî we remove #hashtags, numbers (123), and special chars like @ or $."]
```

```python
len(t)
```

```python
clean_text = []
for i in t:
    clean_text.append(i.lower())
print(clean_text)
```

```python
# removing emojis and specioal characters, and number
def clean_special(text):
    text = re.sub(r'[^a-zA-Z\s]',"",text)
    return text
```

```python
cleaned_text = [clean_special(t) for t in clean_text ]
```

```python
cleaned_text
```

```python
# Tokenize
tokenized_text = []
for i in cleaned_text:
    tokenized_text.append(word_tokenize(i))
```

```python
stop_words = set(stopwords.words("english"))
new_text = []
for i in tokenized_text:
    for x in i:
        if x not in stop_words:
            new_text.append(x)
    # print(i)
```

```python
# Stemming
stemmer = PorterStemmer()
for i in new_text:
    print(i,"<---------------->",stemmer.stem(i))
```

```python
# lemmatize
lem = WordNetLemmatizer()
for i in new_text:
    print(i,"<------------------>",lem.lemmatize(i))
```
